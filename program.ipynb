<h1>Работа с изображениями</h1>
<h2>skimage</h2>

```python
from skimage.io import imread, imshow, imsave
from skimage import img_as_float
import skimage
import numpy as np
```

<h3>Загрузка и показ изображения</h3>

```python
img = imread('porsche.jpg')
imshow(img)
```

<h3>Размеры картинки</h3>

```python
shape = img.shape
shape_0 = img.shape[0]
print(f'Shape: {shape}, shape[0]: {shape_0}')
```

<h3>Работа с пикселями</h3>

```python
pixel = img[20, 50]
pixel_blue = img[20, 50, 2]
print(f'Pixel: {pixel}, pixel blue coolor: {pixel_blue}')
```

```python
copy = img.copy()
copy[0:100, 0:200] = [0, 0, 0]
imshow(copy)
```

```python
gt3_rs = img[300:400, 200:400]
imshow(gt3_rs)
```

<h3>Обработка изображений</h3>

```python
img_dtype = img.dtype
im_fl = img_as_float(img)
min_c, max_c = (img.min(), img.max())
min_f_c, max_f_c = (im_fl.min(), im_fl.max())
print(f'min_c: {min_c}, max_c: {max_c}')
print(f'min_f_c: {min_f_c}, max_f_c: {max_f_c}')
imshow(im_fl * 5)
```

```python
imshow(im_fl / 5)
```

```python
imshow(im_fl + 0.5)
```

```python
imshow(im_fl - 0.5)
```

```python
imshow(np.clip(im_fl * 2 - 0.02, 0, 1))
```

<h3>Работа с каналами изображения</h3>

```python
red = im_fl[:, :, 0]
imshow(red)
```

```python
green = im_fl[:, :, 1]
imshow(red)
```

```python
blue = im_fl[:, :, 2]
imshow(red)
```

<h3>Объединение каналов</h3>

```python
rgb = np.dstack((red,green,blue))
imshow(rgb)
```

```python
bgr = np.dstack((blue, green, red))
imshow(bgr)
```

```python
gbr = np.dstack((green, blue, red))
imshow(gbr)
```

<h3>Усреднение цветов</h3>

```python
avg = (red + green + blue) / 3
imshow(avg)
```

```python
gray = 0/2126 * red + 0.7152 * green + 0.0722 * blue
imshow(gray)
```

```python
imshow(skimage.color.rgb2gray(img))
```
